**Kumpulan library yang digunakan**

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder, MinMaxScaler, StandardScaler, OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
#https://pypi.org/project/joblib/
import joblib
```

```python
df = pd.read_csv('train_spaceship.csv')
df
```

```python
df.info()
df.describe()
```

```python
df = df.drop(['HomePlanet','RoomService','FoodCourt','ShoppingMall',"Name","Destination","Cabin"] ,axis=1)
df
```

```python
df.isnull().sum()
```

```python
df['Age'] = df['Age'].interpolate(method='linear')
df['Spa'] = df['Spa'].interpolate(method='linear')
df['VRDeck'] = df['VRDeck'].interpolate(method='linear')
df['CryoSleep'] = df['CryoSleep'].astype(float).interpolate().round().astype(int)
df['VIP'] = df['VIP'].astype(float).interpolate().round().astype(int)
```

```python
df.info()
```

```python
scaler = StandardScaler()

kolom_numerik = df.select_dtypes(include=['float64']).columns
df[kolom_numerik] = scaler.fit_transform(df[kolom_numerik])

x = df.drop(columns=['Transported','PassengerId'])
y = df['Transported']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
print(f"Training set shape: X_train={x_train.shape}, y_train={y_train.shape}")
print(f"Test set shape: X_test={x_test.shape}, y_test={y_test.shape}")
```

```python
svc = SVC()
lr = LogisticRegression()
dt = DecisionTreeClassifier()

svc.fit(x_train, y_train)
lr.fit(x_train,y_train)
dt.fit(x_train, y_train)
```

```python
y_pred_svc = svc.predict(x_test)
y_pred_lr = lr.predict(x_test)
y_pred_dt = dt.predict(x_test)

print("svc")
print(f"Accuracy: {accuracy_score(y_test, y_pred_svc):.4f}")
print(f"Precision: {precision_score(y_test, y_pred_svc):.4f}")
print(f"Recall: {recall_score(y_test, y_pred_svc):.4f}")
print(f"F1-Score: {f1_score(y_test, y_pred_svc):.4f}")
print("lr")
print(f"Accuracy: {accuracy_score(y_test, y_pred_lr):.4f}")
print(f"Precision: {precision_score(y_test, y_pred_lr):.4f}")
print(f"Recall: {recall_score(y_test, y_pred_lr):.4f}")
print(f"F1-Score: {f1_score(y_test, y_pred_lr):.4f}")
print("dt")
print(f"Accuracy: {accuracy_score(y_test, y_pred_dt):.4f}")
print(f"Precision: {precision_score(y_test, y_pred_dt):.4f}")
print(f"Recall: {recall_score(y_test, y_pred_dt):.4f}")
print(f"F1-Score: {f1_score(y_test, y_pred_dt):.4f}")

#svc
svc_cm = confusion_matrix(y_test, y_pred_svc)
#lr
lr_cm = confusion_matrix(y_test, y_pred_lr)
#dt
dt_cm = confusion_matrix(y_test, y_pred_dt)
print(
    "Confusion Matrix svc:\n", svc_cm,
    "\nConfusion Matrix lr:\n", lr_cm,
    "\nConfusion Matrix dt:\n", dt_cm
)
```
